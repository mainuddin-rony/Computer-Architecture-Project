```python
size_mapper = {'128kB': 128, '256kB': 256, '512kB': 512, '1MB': 1024, '2MB': 2048}
```

```python
import os

req_lines = [360, 368, 376, 384, 392, 400, 465, 471, 477, 483, 489, 495, 595, 596, 597, 606, 607, 608, 617, 618, 619, 630, 631, 632, 641, 642, 643, 652, 653, 654]
value_list =[]
for filename in os.listdir("/home/mrony/gem5 stats"):
    if filename.endswith('.txt'):
        f = open(filename, 'r')
        lines = f.readlines()
        lines = lines[2:len(lines)-2]
        value = {}
        for line in lines:
            values = line.split()
            if values[0] in mapper:
                params = filename.split("_")
                value['benchmark'] = params[0]
                value['l1d_size'] = params[1]
                value['l1i_size'] = params[2]
                value['l2_size'] = size_mapper[params[3]]
                value['l2_type'] = params[4].split(".")[0]
                value[mapper[values[0]]] = float(values[1])
        
        value_list.append(value)
#         for i in req_lines:
#             value[mapper[lines[i].split()[0]]] = lines[i].split()[1]
#         value_list.append(value)
        
```

```python
all_stats = pd.DataFrame(value_list)
```
